## Advanced Topics

### Batched planning and execution

```python
from robotodo.engines.isaac.articulation import Articulation

panda_batch = Articulation.load(
    "/World/Panda{0..10}",
    "https://omniverse-content-production.s3-us-west-2.amazonaws.com/Assets/Isaac/5.0/Isaac/Robots/FrankaRobotics/FrankaPanda/franka.usd",
    scene=scene,
)
panda_batch
```

```python
import numpy
from robotodo.utils.pose import Pose

# TODO NOTE ensure sufficient space between robots to avoid collisions
panda_batch.pose = Pose(
    p=numpy.linspace([0, 1, 0], [0, 25, 0], num=len(panda_batch.path))
)
```

```python
planner_batch = panda_batch.planner(
    base_link="panda_link0",
    end_link="panda_hand",
)
```

```python
from robotodo.utils.pose import Pose

# TODO NOTE start the simulation so that the robot can move
async with scene.play():
    await panda_batch.driver.execute_action(
        planner_batch.compute_action({"target_pose_from_base": Pose([.5, 0, .5])}),
        position_error_limit=1,
        velocity_error_limit=1,
    )
```
