```python
controller.action_spec
```

```python
# call .reshape here because the dimensions are flexible
# TODO upstream tensorspecs: allow size to be BoxSpec so that the range can be specified?
observation = controller.observation_spec.reshape({
    # "batch?": 10,
    "height": 32,
    "width": 32,
    "channels": 3,
}).random()
observation
```

```python
controller.compute_action(observation)
```

```python
action = controller.action_spec.reshape({
    # "batch?": 10,
    "timestep": 5,
}).random()
action
```

```python
controller.learn(observation, action, training_num_steps=1)
```
