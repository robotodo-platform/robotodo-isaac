```python
import asyncio
import numpy as np
import matplotlib.pyplot as plt

import omni.kit.app
import omni.appwindow
from omni.kit.renderer import bind as okr
import omni.kit.renderer_capture as rcap

app = omni.kit.app.get_app()
win = omni.appwindow.get_default_app_window()
renderer = okr.get_renderer_interface()
cap = rcap.acquire_renderer_capture_interface()

# keep capture async (default). If you prefer a blocking capture, set True.
# cap.set_capture_sync(False, win)

todo = ...

async def capture_app_window_once(timeout_s: float = 2.0) -> np.ndarray:
    """
    Capture the *next presented frame* of the WHOLE app window (swapchain)
    and return it as an HxWx4 uint8 RGBA NumPy array.
    """
    loop = asyncio.get_running_loop()
    fut: asyncio.Future = loop.create_future()

    # <capsule object NULL at 0x7f5a56d29740> 5184000 1440 900 TextureFormat.BGRA8_UNORM
    def _on_bytes(buf_capsule, size, width, height, tex_format):
        print(buf_capsule, size, width, height, tex_format)

        # Convert raw swapchain bytes -> RGBA tuples -> NumPy (HxWx4, uint8)        
        rgba_list = rcap.convert_raw_bytes_to_rgba_tuples(
            buf_capsule, size, width, height, tex_format
        )
        arr = np.asarray(rgba_list, dtype=np.uint8).reshape(height, width, 4)
        if not fut.done():
            fut.set_result(arr)

    # def _on_bytes(*args):
    #     global todo
    #     todo = args
    #     print(args)

    # Request: capture the NEXT frame and call our callback with raw bytes
    cap.capture_next_frame_swapchain_callback(_on_bytes, app_window=win)

    # Drive frames until the capture fires (works in notebooks too)
    while not fut.done():
        # advance Kit one frame; resolves after the next present
        await app.next_update_async()
    return await fut

# ---- run once and visualize with matplotlib ----
arr = await capture_app_window_once()

plt.figure(figsize=(8, 6))
# matplotlib accepts RGBA; if you want RGB only, use arr[..., :3]
plt.imshow(arr)
plt.axis("off")
plt.title("Current App Window (Swapchain)")
plt.show()
```
